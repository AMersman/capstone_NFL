```python
'''Change data types to Int64'''
stadiums.stadium_capacity = stadiums.stadium_capacity.replace(',','', regex=True)
stadiums.stadium_capacity = stadiums.stadium_capacity.astype("Int64")
stadiums.stadium_open = stadiums.stadium_open.astype("Int64")
stadiums.stadium_close = stadiums.stadium_close.astype("Int64")
```


The solar azimuth angle is the azimuth (horizontal angle with respect to north) of the Sun's position. This horizontal coordinate defines the Sun's relative direction along the local horizon, whereas the solar zenith angle (or its complementary angle solar elevation) defines the Sun's apparent altitude.

```python
'''Create column for game winner'''
#establish who won each game
winners = []

#iterate through games compariong scores to determine winners, append the winner to the list
for i,v in games.score_home.items():
    if games.score_home[i] > games.score_away[i]:
        winners.append('Home')
        
    elif games.score_away[i]>games.score_home[i]:
        winners.append('Away')
        
    else:
        winners.append("Tie")
        
#convert list to column                       
games["winner"] = winners
                       

        
```

```python
'''Create column for game winner'''
#establish who won each game
losers = []

#iterate through games compariong scores to determine winners, append the winner to the list
for i,v in games.score_home.items():
    if games.score_home[i] < games.score_away[i]:
        losers.append('Home')
        
    elif games.score_away[i]<games.score_home[i]:
        losers.append('Away')
        
    else:
        losers.append("Tie")
        
#convert list to column                       
games["loser"] = losers
```

```python
'''Change team_favorite_id to full name to match the team_home and team_away columns'''
games['team_favorite_id'] = games.team_favorite_id.map(teams.set_index('team_id')['team_name'].to_dict())
```

```python
'''Verify changes'''
games[games['team_favorite_id'].notnull()]
```

```python
games.schedule_week.unique()
```

```python
'''Change Text values of schedule week to int for future use in correlation matrix'''

week_mapping = {
    'Wildcard': 19,
    'Division': 20,
    'Conference': 21,
    'Superbowl': 22
}
# Replace values in the schedule_week column using the mapping
games['schedule_week'] = games['schedule_week'].replace(week_mapping)
# Convert the schedule_week column to a numeric data type (integer)
games['schedule_week'] = pd.to_numeric(games['schedule_week'], errors='coerce')
# Sort the dataframe in ascending order by schedule_season and schedule_week
sorted_games = games.sort_values(by=['schedule_season', 'schedule_week'], ascending=True)
```

```python
games.schedule_week.unique()
```

```python
'''Creation of columns for score difference and total and correction of spread favorite 
to match standard understanding of spread'''

games['score_difference'] = games['score_away'] - games['score_home']
games['score_total'] = games['score_away'] + games['score_home']

games['spread_favorite'] = np.where(games['team_favorite_id'] == games['team_away'],
                                       np.abs(games['spread_favorite']), games['spread_favorite'])
```

```python
'''Verify the use of weather_detail to contain info about precipitation'''
games.weather_detail.unique().tolist()
```

```python
'''Change null values to indicate no precipitation'''
games['weather_detail'].fillna('No Precip', inplace = True)
```

```python
'''Merge stadiums and games dataframes on stadium name'''

merged_df = games.merge(stadiums, left_on="stadium",
                       right_on="stadium_name", suffixes=['_x', '_y'], how="left")

merged_df.info()
```

```python
'''Select desired columns and fill in null values using appropriate method'''

merged_df = merged_df[['schedule_date', 'schedule_season', 'schedule_week', 'schedule_playoff', 
                       'team_home', 'team_away', 'score_home', 'score_away', 'winner', 'loser', 'spread_favorite', 'score_difference', 'score_total'
                       ,'stadium', 'stadium_neutral', 'stadium_type', 'stadium_elevation',
                      'weather_temperature', 'weather_wind_mph', 'weather_detail', 'weather_humidity']]

merged_df = merged_df.dropna(subset=['score_home'])
merged_df = merged_df.dropna(subset=['stadium_type'])
merged_df = merged_df.dropna(subset=['spread_favorite'])

mean_humidity = merged_df['weather_humidity'].mean()
merged_df["weather_humidity"].fillna(mean_humidity, inplace = True)

mean_temp = merged_df['weather_temperature'].mean()
merged_df["weather_temperature"].fillna(mean_temp, inplace = True)

mean_wind = merged_df['weather_wind_mph'].mean()
merged_df["weather_wind_mph"].fillna(mean_wind, inplace = True)



mean_elevation = merged_df['stadium_elevation'].mean()
merged_df["stadium_elevation"].fillna(mean_elevation, inplace = True)



```
